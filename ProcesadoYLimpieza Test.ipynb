```python
import pandas as pd
import numpy as np
```

```python
CatEmisora = pd.read_csv("ExtraccionDomiVersionFinal\CatEmisora.csv")
CatBanco = pd.read_csv("ExtraccionDomiVersionFinal\CatBanco.csv")
EmisoraBancoPrecios = pd.read_csv("ExtraccionDomiVersionFinal\EmisoraBancoPrecios.csv")
ListaCobroEmisora = pd.read_csv("ExtraccionDomiVersionFinal\ListaCobroEmisora.csv")
CatRespuestaBancos = pd.read_csv("ExtraccionDomiVersionFinal\CatRespuestaBancos.csv")
ListaCobroYYYY = pd.read_csv("ExtraccionDomiVersionFinal/ListaCobro.csv")
```

```python
# Read and concatenate all ListaCobroDetalle CSVs
lista_cobro_files = [
    "ExtraccionDomiVersionFinal\\ListaCobroDetalle2025.csv"
]

ListaCobroDetalle = pd.concat([pd.read_csv(f) for f in lista_cobro_files], ignore_index=True)
```

```python
# Engels
ListaCobroEmisoraCompleta = pd.merge(EmisoraBancoPrecios, ListaCobroEmisora, on="idEmisora", suffixes=["","_Emisora"])
ListaCobroDetalleCompleta = pd.merge(ListaCobroDetalle, CatBanco, left_on="idBanco", right_on="IdBanco").drop("idBanco", axis=1)
ListaCobroDetalleCompleta = pd.merge(ListaCobroDetalleCompleta, ListaCobroEmisoraCompleta, on="idListaCobro", suffixes=["_Credito","_Emisora"])
ListaCobroDetalleCompleta.fillna({"idRespuestaBanco": 33.0}, inplace=True)
ListaCobroDetalleCompleta.idRespuestaBanco = ListaCobroDetalleCompleta.idRespuestaBanco.apply(lambda x: str(int(x)).zfill(2) if type(x) is float else x)
ListaCobroDetalleCompleta = pd.merge(ListaCobroDetalleCompleta, CatRespuestaBancos, left_on="idRespuestaBanco", right_on="IdRespuestaBanco").drop("idRespuestaBanco", axis=1)
ListaCobroDetalleCompleta = pd.merge(ListaCobroDetalleCompleta, ListaCobroYYYY,  on="idListaCobro", suffixes=["fechaEnvioCobro"])
```

```python
ListaCobroDetalleCompleta["transCount"] = (ListaCobroDetalleCompleta.sort_values(by=["idCredito", "consecutivoCobro"]).groupby("idCredito").cumcount()+1)
```

```python
#ListaCobroDetalleCompleta["transSuccess"]  
# Aplicamos el conteo solo sobre las filas con fecha vÃ¡lida
mask_valid_date = ListaCobroDetalleCompleta['fechaCobroBanco'].notna()
ListaCobroDetalleCompleta['transSuccess'] = 0 # Inicializamos con NaN
ListaCobroDetalleCompleta.loc[mask_valid_date, 'transSuccess'] = (
    ListaCobroDetalleCompleta[mask_valid_date]
    .sort_values(by=['idCredito', 'consecutivoCobro'])
    .groupby('idCredito')
    .cumcount() + 1 
)
```

```python
# Asegurarse de float
ListaCobroDetalleCompleta['transSuccess'] = ListaCobroDetalleCompleta['transSuccess'].astype(float)

# Hacer copia temporal
temp = ListaCobroDetalleCompleta.copy()

# Crear una mÃ¡scara para detectar la primera fila de cada idCredito
first_in_group_mask = temp.groupby("idCredito").cumcount() == -1

# Solo convertir a NaN los ceros que NO son la primera apariciÃ³n del grupo
temp.loc[(temp['transSuccess'] == 0) & (~first_in_group_mask), 'transSuccess'] = np.nan

# Forward-fill dentro de cada idCredito
temp['transSuccess'] = temp.groupby('idCredito')['transSuccess'].ffill()

# Restaurar ceros donde no se haya podido rellenar
temp['transSuccess'] = temp['transSuccess'].fillna(0)

# Guardar en el DataFrame original
ListaCobroDetalleCompleta['transSuccess'] = temp['transSuccess']
ListaCobroDetalleCompleta["transResidual"] = ListaCobroDetalleCompleta['transCount'] - ListaCobroDetalleCompleta['transSuccess'] 
```

```python
ListaCobroDetalleCompleta = ListaCobroDetalleCompleta.sort_values(by=['idCredito', 'consecutivoCobro'])  # asegurarse del orden temporal

ListaCobroDetalleCompleta['pagoAnterior'] = ListaCobroDetalleCompleta.groupby('idCredito')['montoCobrado'].shift(1)
ListaCobroDetalleCompleta['pagoAnterior'] = ListaCobroDetalleCompleta['pagoAnterior'].fillna(0.0)
```

```python
# Assuming your DataFrame is already loaded
df = ListaCobroDetalleCompleta.copy()

# --- 1. Ratios and Residuals ---
df['ratio_cobrado_cobrar'] = df['montoCobrado'] / df['montoCobrar'].replace(0, np.nan)
df['residual_cobrar'] = df['montoCobrar'] - df['montoCobrado']

df["Costo"] = 0
df['Costo'] = np.where(df['fechaCobroBanco'].isna(), df['Costo_Hit_Miss'], df['Costo_Hit_Win'])

df['residualAnterior'] = df.groupby('idCredito')['montoCobrado'].shift(1)
df['residualAnterior'] = df['residualAnterior'].fillna(0.0)
df['ratioAnterior'] = df.groupby('idCredito')['ratio_cobrado_cobrar'].shift(1)
df['ratioAnterior'] = df['ratioAnterior'].fillna(0.0)

# Days since last payment per loan
#df = df.sort_values(by=['idCredito', 'fechaCobroBanco'])
#df['days_since_last_payment'] = df.groupby('idCredito')['fechaCobroBanco'].diff().dt.days

# --- 3. Attempt Count (per loan) ---
#df['max_consecutivoCobro'] = df.groupby('idCredito')['consecutivoCobro'].transform('max')

# Preview
df = df.drop(columns = ["montoExigible", "Nombre_Credito", "NombreEmisora", "Emisora", "Nombre_Emisora", "fechaCreacionLista", "Costo_Hit_Miss", "Costo_Hit_Win"])
df.head()
```
